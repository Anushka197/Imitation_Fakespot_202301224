### Fake Reviews

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../data/processed/cleaned_reviews.csv")

fake_reviews = df[df['label'] == 'OR']['cleaned_review']

text = " ".join(fake_reviews)

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white'
).generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud for Fake Reviews")
plt.show()
```

### Original Reviews

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd

# Load data
df = pd.read_csv("../data/processed/cleaned_reviews.csv")

# Filter for 'CG' (Genuine Reviews)
# Added .dropna() to ensure no errors if some rows are empty
genuine_reviews = df[df['label'] == 'CG']['text_'].dropna().astype(str)

text = " ".join(genuine_reviews)

# Generate WordCloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    colormap='Greens' # Green is a good visual cue for 'Genuine/Safe'
).generate(text)

# Plotting
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud for Genuine Reviews (Consumer Generated)")
plt.show()
```
